<h1>Vegetation data analysis</h1>

```python
import pandas as pd
import numpy as np 
import plotly.express as px
```

```python
#Grenness fraction data
path_gf_data = "C:/pale-blue-dot-challenge/data/g4.areaAvgTimeSeries.M2TMNXLND_5_12_4_GRN.19800101-20231130.180W_90S_180E_90N.csv"
gf_data = pd.read_csv(path_gf_data, skiprows=9,  usecols=[0,1], names=['Time','Greenness fraction'])
```

```python
gf_data.head()
```

```python
gf_data.tail()
```

```python
gf_fig = px.line(gf_data, x="Time", y="Greenness fraction")
gf_fig.show()
```

```python
gf_fig.update_traces(line=dict(color='green'))
gf_fig.show()
```

```python
# Leaf area index data
path_lai_data = "C:/pale-blue-dot-challenge/data/g4.areaAvgTimeSeries.M2TMNXLND_5_12_4_LAI.19800101-20231130.180W_90S_180E_90N.csv"
lai_data = pd.read_csv(path_lai_data, skiprows=9,  usecols=[0,1], names=['Time','Leaf area index'])
```

```python
lai_fig = px.line(lai_data, x="Time", y="Leaf area index")
lai_fig.update_traces(line=dict(color='green'))
lai_fig.show()
```
